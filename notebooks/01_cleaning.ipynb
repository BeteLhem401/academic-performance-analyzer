# Student Success Factors Dashboard — Data Cleaning


```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

## Load Dataset

Load raw survey responses to inspect student attendance, study habits, and GPA.


```python
df = pd.read_csv('../data/raw/raw_survey_responses.csv')
df.head()
```

## Dataset Overview

Check the dataset shape and column types to understand its structure and quality.


```python
print(df.shape)
df.info()
```

## Missing Values

Identify missing values across columns and evaluate which gaps need filling.


```python
df.isnull().sum()
```

```python
# Replace missing or empty 'main_challenge' values with "Unknown"
if 'df' in globals() or 'df' in locals():
    try:
        if 'main_challenge' in df.columns:
            df['main_challenge'] = df['main_challenge'].fillna('Unknown')
            df['main_challenge'] = df['main_challenge'].replace('', 'Unknown')
            print("Filled missing 'main_challenge' values with 'Unknown'.")
        else:
            print("Column 'main_challenge' not found in `df`.")
    except Exception as e:
        print('Error filling main_challenge values:', e)
else:
    print("DataFrame `df` not defined yet. Run the earlier cells first.")
```

Only the `main_challenge` column contains missing values, while all other columns are complete.

This suggests most students completed the survey carefully.

The missing challenge responses may indicate students skipped sensitive or unclear questions.

## Remove Duplicates

Remove duplicate survey entries so the cleaned dataset reflects unique students.


```python
df = df.drop_duplicates()
print(df.shape)
```

## Standardize Text

Trim whitespace from text fields for consistent categorical values.


```python
text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = df[col].astype(str).str.strip()
```

## Convert Numeric Columns

Convert key survey fields to numeric data so metrics can be analyzed reliably.


```python
numeric_columns = [
    'attendance_percent',
    'study_hours_week',
    'assignment_completion_rate',
    'gpa',
    'course_difficulty',
    'sleep_hours_night',
    'mental_health_rating'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

Numerical columns were converted into proper numeric data types for calculations and visualization.


### Suggestion

Numeric validation should be applied directly inside the survey form.


## Handle Missing Values

Fill missing numeric values with medians to keep student metrics complete.


```python
for col in numeric_columns:
    df[col] = df[col].fillna(df[col].median())
```

## Validate Ranges

Fix invalid attendance and GPA values so the dataset contains realistic student records.


```python
df.loc[(df['attendance_percent'] < 0) | (
    df['attendance_percent'] > 100), 'attendance_percent'] = np.nan
df.loc[(df['gpa'] < 0) | (df['gpa'] > 4.0), 'gpa'] = np.nan

df['attendance_percent'] = df['attendance_percent'].fillna(
    df['attendance_percent'].median())
df['gpa'] = df['gpa'].fillna(df['gpa'].median())
```

## Feature Engineering

Create academic risk labels and engagement scores to summarize student success factors.


```python
df['academic_risk'] = np.where(
    df['gpa'] < 2.5,
    'High Risk',
    np.where(df['gpa'] < 3.2, 'Moderate Risk', 'Low Risk')
)

df['engagement_score'] = (
    (df['attendance_percent'] * 0.4) +
    (df['assignment_completion_rate'] * 0.4) +
    (df['study_hours_week'] * 2)
)

df[['anonymous_id', 'academic_risk', 'engagement_score']].head()
```

## Export Clean Dataset

Save the cleaned survey data so downstream analysis and dashboarding use reliable records.


```python
df.to_csv('../data/processed/clean_survey_responses.csv', index=False)
print('Dataset exported successfully.')
```
